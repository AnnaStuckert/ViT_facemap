```python
df_demo = pd.read_hdf(collected_data_files[1])
df_demo.head()
```

```python
bodyparts = df_demo.columns.get_level_values("bodyparts").unique().to_list()
bodyparts
```

```python
def merge_annotateddatasets(dataset_h5_files: list, bodyparts: list):
    annotation = []

    for file_path in dataset_h5_files:
        data = pd.read_hdf(file_path)
        conversioncode.guarantee_multiindex_rows(data)
        annotation.append(data)
    
    annotation = pd.concat(annotation).sort_index()
    annotation = annotation.reindex(
        bodyparts,
        axis=1,
        level=annotation.columns.names.index("bodyparts"),
    )
    return annotation


df_merged = merge_annotateddatasets(collected_data_files, bodyparts)
df_merged
```

```python
scorer = "AVS"
filename = f"CollectedData_{scorer}"

df_merged.to_hdf(filename + ".h5", key="keypoints", mode="w")
df_merged.to_csv(filename + ".csv")
```

```python
df_fixed = df_merged.copy()
df_fixed.index = pd.MultiIndex.from_tuples(
    [
        ("labeled-data", video_folder, image_name)
        for video_folder, subfolder, image_name in df_fixed.index
    ]
)
df_fixed
```

```python
scorer = "AVS"
filename = f"CollectedData_{scorer}"

df_fixed.to_hdf(filename + ".h5", key="keypoints", mode="w")
df_fixed.to_csv(filename + ".csv")
```
